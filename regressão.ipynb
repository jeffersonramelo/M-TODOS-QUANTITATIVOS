```python
import pandas as pd

df = pd.read_excel('/content/dados regressão multipla_Revisão.xlsx')

print(df.head())
```

```python
# Corrected definition of independent (X) and dependent (y) variables
X = df[['divida']]
y = df['Lucro']
```

```python
import statsmodels.api as sm

# Add a constant to the independent variable for statsmodels
X_sm = sm.add_constant(X)

# Create and fit the OLS (Ordinary Least Squares) model
model_sm = sm.OLS(y, X_sm)
results_sm = model_sm.fit()

# Print the summary in a Stata-like format
print(results_sm.summary())
```
