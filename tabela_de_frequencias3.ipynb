```python
import pandas as pd

# Lendo o arquivo CSV
df = pd.read_csv('/content/exercicio03frequencias.csv', sep=';')

# Exibindo as primeiras linhas do DataFrame
df
```

Tabela de frequências para variáveis QUALITATIVAS e  QUANTITATIVAS_SEM ORGANIZAÇÃO EM CLASSES

---





```python
import pandas as pd

# Supondo que seu DataFrame seja 'df' e já foi carregado com pd.read_csv()

# Frequência Simples (FS)
fs = df['porte'].value_counts()

# Frequência Acumulada (FA)
fa = fs.cumsum()

# Frequência Relativa Simples (FRS) - Convertendo para porcentagem
frs = (fs / fs.sum()) * 100

# Frequência Relativa Acumulada (FRA)
fra = frs.cumsum()

# Criando um novo DataFrame para a tabela de distribuição de frequência
tabela_frequencia = pd.DataFrame({
    'Frequência Simples': fs,
    'Frequência Acumulada': fa,
    'Frequência Relativa Simples (%)': frs,
    'Frequência Relativa Acumulada (%)': fra
})

# Resetando o índice para incluir a variável 'sexo' como uma coluna
tabela_frequencia = tabela_frequencia.reset_index().rename(columns={'index': 'porte'})

print(tabela_frequencia)
```

Salvando a tabela em Excel

```python
# Continuando do código anterior após a criação da tabela_frequencia

# Salvando a tabela em um arquivo Excel
tabela_frequencia.to_excel('distribuicao_frequencia_qualificacao.xlsx', index=False, engine='openpyxl')
```

Tabela de frequências para variáveis QUANTITATIVAS_COM ORGANIZAÇÃO EM CLASSES

```python
# Encontrando o valor minimo na coluna 'renda'
valor_minimo_renda = df['lucro'].min()
valor_minimo_renda
```

```python
# Encontrando o valor máximo na coluna 'renda'
valor_maximo_renda = df['lucro'].max()
valor_maximo_renda
```

```python
import pandas as pd

# Supondo que seu DataFrame seja 'df' e já foi carregado com pd.read_csv()

# Definindo as classes de renda
bins = [1000, 20000, 40000, 60000, 80499]  # O último intervalo vai de 800 até infinito
labels = ['1000 a 20000', '20001 a 400000', '40001 a 600000', '60001 a 80499']
df['Categoria lucro'] = pd.cut(df['lucro'], bins=bins, labels=labels, right=False)

# Frequência Simples (FS) para 'Categoria lucro'
fs_gestao = df['Categoria lucro'].value_counts().sort_index()

# Frequência Acumulada (FA) para 'Categoria lucro'
fa_gestao = fs_gestao.cumsum()

# Frequência Relativa Simples (FRS) - Convertendo para porcentagem
frs_gestao = (fs_gestao / fs_gestao.sum()) * 100

# Frequência Relativa Acumulada (FRA)
fra_gestao = frs_gestao.cumsum()

# Criando um novo DataFrame para a tabela de distribuição de frequência para 'lucro'
tabela_frequencia_gestao = pd.DataFrame({
    'Categoria lucro': labels,
    'Frequência Simples': fs_gestao,
    'Frequência Acumulada': fa_gestao,
    'Frequência Relativa Simples (%)': frs_gestao,
    'Frequência Relativa Acumulada (%)': fra_gestao
})

print(tabela_frequencia_gestao)
```

```python
# Continuando do código anterior após a criação da tabela_frequencia

# Salvando a tabela em um arquivo Excel
tabela_frequencia.to_excel('distribuicao_frequencia_renda.xlsx', index=False, engine='openpyxl')
```
