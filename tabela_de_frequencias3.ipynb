```python
import pandas as pd

# Lendo o arquivo CSV
df = pd.read_csv('/content/exercicio03frequencias.csv', sep=';')

# Exibindo as primeiras linhas do DataFrame
df
```

Tabela de frequÃªncias para variÃ¡veis QUALITATIVAS e  QUANTITATIVAS_SEM ORGANIZAÃ‡ÃƒO EM CLASSES

---





```python
import pandas as pd

# Supondo que seu DataFrame seja 'df' e jÃ¡ foi carregado com pd.read_csv()

# FrequÃªncia Simples (FS)
fs = df['porte'].value_counts()

# FrequÃªncia Acumulada (FA)
fa = fs.cumsum()

# FrequÃªncia Relativa Simples (FRS) - Convertendo para porcentagem
frs = (fs / fs.sum()) * 100

# FrequÃªncia Relativa Acumulada (FRA)
fra = frs.cumsum()

# Criando um novo DataFrame para a tabela de distribuiÃ§Ã£o de frequÃªncia
tabela_frequencia = pd.DataFrame({
    'FrequÃªncia Simples': fs,
    'FrequÃªncia Acumulada': fa,
    'FrequÃªncia Relativa Simples (%)': frs,
    'FrequÃªncia Relativa Acumulada (%)': fra
})

# Resetando o Ã­ndice para incluir a variÃ¡vel 'sexo' como uma coluna
tabela_frequencia = tabela_frequencia.reset_index().rename(columns={'index': 'porte'})

print(tabela_frequencia)
```

Salvando a tabela em Excel

```python
# Continuando do cÃ³digo anterior apÃ³s a criaÃ§Ã£o da tabela_frequencia

# Salvando a tabela em um arquivo Excel
tabela_frequencia.to_excel('distribuicao_frequencia_qualificacao.xlsx', index=False, engine='openpyxl')
```

Tabela de frequÃªncias para variÃ¡veis QUANTITATIVAS_COM ORGANIZAÃ‡ÃƒO EM CLASSES

```python
# Encontrando o valor minimo na coluna 'renda'
valor_minimo_renda = df['lucro'].min()
valor_minimo_renda
```

```python
# Encontrando o valor mÃ¡ximo na coluna 'renda'
valor_maximo_renda = df['lucro'].max()
valor_maximo_renda
```

```python
import pandas as pd

# Supondo que seu DataFrame seja 'df' e jÃ¡ foi carregado com pd.read_csv()

# Definindo as classes de renda
bins = [1000, 20000, 40000, 60000, 80499]  # O Ãºltimo intervalo vai de 800 atÃ© infinito
labels = ['1000 a 20000', '20001 a 400000', '40001 a 600000', '60001 a 80499']
df['Categoria lucro'] = pd.cut(df['lucro'], bins=bins, labels=labels, right=False)

# FrequÃªncia Simples (FS) para 'Categoria lucro'
fs_gestao = df['Categoria lucro'].value_counts().sort_index()

# FrequÃªncia Acumulada (FA) para 'Categoria lucro'
fa_gestao = fs_gestao.cumsum()

# FrequÃªncia Relativa Simples (FRS) - Convertendo para porcentagem
frs_gestao = (fs_gestao / fs_gestao.sum()) * 100

# FrequÃªncia Relativa Acumulada (FRA)
fra_gestao = frs_gestao.cumsum()

# Criando um novo DataFrame para a tabela de distribuiÃ§Ã£o de frequÃªncia para 'lucro'
tabela_frequencia_gestao = pd.DataFrame({
    'Categoria lucro': labels,
    'FrequÃªncia Simples': fs_gestao,
    'FrequÃªncia Acumulada': fa_gestao,
    'FrequÃªncia Relativa Simples (%)': frs_gestao,
    'FrequÃªncia Relativa Acumulada (%)': fra_gestao
})

print(tabela_frequencia_gestao)
```

```python
# Continuando do cÃ³digo anterior apÃ³s a criaÃ§Ã£o da tabela_frequencia

# Salvando a tabela em um arquivo Excel
tabela_frequencia.to_excel('distribuicao_frequencia_renda.xlsx', index=False, engine='openpyxl')
```
